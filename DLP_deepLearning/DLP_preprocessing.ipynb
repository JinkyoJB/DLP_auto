### crop 꼭지점 구하기

```python
def con_area(contours):
    contour_list = []
    for i in range(len(contours)):
        if cv2.contourArea(contours[i])>10000:
            contour_list.append(i)
    return contour_list
```

```python
def make_point(contour_list):
    min_x_list=[]
    min_y_list=[]
    max_x_list=[]
    max_y_list=[]
    for i in contour_list :
            contour = np.array(contours[i])
            
            min_v = np.min(contour, axis=0)
            min_x = min_v[0][0]
            min_y = min_v[0][1]
            min_x_list.append(min_x)
            min_y_list.append(min_y)

            max_v = np.max(contour, axis=0)
            max_x = max_v[0][0]
            max_y = max_v[0][1]
            max_x_list.append(max_x)
            max_y_list.append(max_y)
    return (min_x_list, min_y_list, max_x_list, max_y_list)
```

```python
pole_point = make_point(contour_list)
print(pole_point)
```

### 좌표로 crop, resize후 저장

네모경화면으로 crop한 이미지를 
'(출력일자)_(레진종류)_(적층 층)' 명으로 '경화표면 이미지 폴더'저장

```python
def makecrop1(img_dir):
    make_ad = f'./DLP_deepLearning/3_cropped/{target}_cropped'
    makedir(make_ad)
    for (path,dirs,files) in os.walk(img_dir):
        for file in files:
            img_open = cv2.imread(f'{img_dir}/{file}', cv2.IMREAD_COLOR)
            img_target = img_open.copy()
            cv2.imwrite(f'{make_ad}/{file}',img_target) #크롭된 전면 저장하기
            
            for i in range(len(contour_list)):
                crop = img_target[pole_point[1][i]:pole_point[3][i],pole_point[0][i]:pole_point[2][i]]
                crop = cv2.resize(crop, (512,512), interpolation =cv2.INTER_AREA)
                cv2.imwrite(f'{make_ad}/{file}_{i+1}.jpg', crop)
```

```python
def movefile(class_dir):
    now = time.strftime('%y%m%d')
    folder = f"C:/jinkyo_coding/DLP_deepLearning/{now}_dataset(class)"
    makedir(folder)
    folder1 = f'{folder}/{classes[0]}/'
    folder2 = f'{folder}/{classes[1]}/'
    folder3 = f'{folder}/{classes[2]}/'
    folder4 = f'{folder}/{classes[3]}/'
    folder5 = f'{folder}/{classes[4]}/'
    folder6 = f'{folder}/left/'
#     makedir(folder1)
#     makedir(folder2)
#     makedir(folder3)
#     makedir(folder4)
#     makedir(folder5)
#     makedir(folder6)
    
    for (path,dirs,files) in os.walk(class_dir):
        for file in files:
            lable = file[-5]
            if lable == 'N':
                shutil.move(class_dir + '/' + file, folder1 + file)
            elif lable == 'S':
                shutil.move(class_dir + '/' + file, folder2 + file)
            elif lable == 'C':
                shutil.move(class_dir + '/' + file, folder3 + file)
            elif lable == 'X':
                shutil.move(class_dir + '/' + file, folder4 + file)
            elif lable == 'M':
                shutil.move(class_dir + '/' + file, folder5 + file)
            else:
                shutil.move(class_dir + '/' + file, folder6 + file)
```

```python
normal= 'C:/jinkyo_coding/DLP_deepLearning/5_220824_dataset(class)/normal'
sub_normal= 'C:/jinkyo_coding/DLP_deepLearning/5_220824_dataset(class)/sub_normal'
critical= 'C:/jinkyo_coding/DLP_deepLearning/5_220824_dataset(class)/critical'
pore= 'C:/jinkyo_coding/DLP_deepLearning/5_220824_dataset(class)/pore'
minor_defect= 'C:/jinkyo_coding/DLP_deepLearning/5_220824_dataset(class)/minor_defect'

add_list = [normal, sub_normal, critical, pore, minor_defect]
```

```python
f = open('output.csv', 'w', encoding='utf-8', newline='')
writer = csv.writer(f)
for i in add_list:
    for (path,dirs,files) in os.walk(i):
        for file in files:
            name = file
            if i == normal:
                writer.writerow([name, 0])
            elif i == sub_normal:
                writer.writerow([name, 1])
            elif i == critical:
                writer.writerow([name, 2])
            elif i == pore:
                writer.writerow([name, 3])
            elif i == minor_defect:
                writer.writerow([name, 4])
                
f.close()
```
